### Make a histogram on the Price column. Replace function.  
The problem is that the Price column is not numeric, but in can be converted to numeric with replace function

```python
car_sales.dtypes
```

```python
car_sales
```

```python
# .astype(float): this handles values with cents, e.g., $9,700.99 becomes 9700.99
# .round(): rounds to the nearest whole number  15001.0
# .astype(int): finally converts to integer by truncating the decimal (not rounding).
car_sales2 = car_sales.copy()
car_sales2["Price_Clean"] = car_sales2["Price"].str.replace('[\$,]', '', regex=True).astype(float).astype(int)
car_sales2
```

### Import car-sales-missing-data.csv

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

### Fill out the missing cells in Odometer with the mean of Odometer values. Round the mean value to 1 decimal place.

```python
car_sales_missing
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(round(car_sales_missing["Odometer"].mean(), 1))
```

```python
car_sales_missing
```

### Remove all the rows with missing values

```python
car_sales_missing_dropped= car_sales_missing.dropna()
```

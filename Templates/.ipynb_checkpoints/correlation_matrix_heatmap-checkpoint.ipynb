```python
import pandas as pd
import numpy as np
import plotly.express as px

df_boats = pd.read_csv('/home/zolcs/JupiterNotebooks/machine_learning/Statistics/boat24_five_page_ships_cleaned.csv')  

# keeping only numerical variables  
numerical_columns = df_boats.select_dtypes(include=[np.number]).columns  

# the correlation matrix 
corr_matrix = df_boats[numerical_columns].corr()  
corr_matrix

fig = px.imshow(corr_matrix, text_auto=True, aspect="auto",
                labels=dict(color='Correlation'),
                x=corr_matrix.columns,
                y=corr_matrix.columns,
                width = 1000, height = 600,
                title="Correlation Matrix Heatmap")

# Setting the extreme values for [-1;1]
fig.update_coloraxes(cmin=-1, cmax=1)

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate "x" values
np.random.seed(42)  # for reproducibility
x = np.linspace(0, 100, 100)

# Generate "y" with a linear connection + noise
slope = 2.5
intercept = 10
noise = np.random.normal(0, 20, size=x.shape)  # add random variation
y = slope * x + intercept + noise

# Scatter plot

plt.figure(figsize=(8,6), facecolor="lightgray")  # whole canvas background
#ax.set_facecolor("beige")  # plot area bg

plt.scatter(x, y, alpha=0.7, edgecolor='k', color='green')
plt.title("Scatter Plot with Linear Trend + Random Noise")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

## Bar chart

 In a bar chart the x-axis represents **different categories** of a factor variable. The bars are shown separate from one another.

```python
import matplotlib.pyplot as plt

# Example data
categories = ["Category A", "Category B", "Category C", "Category D", "Category E"]
values = [23, 45, 12, 36, 29]

# Define colors for bars
colors = ["#4C72B0", "#55A868", "#C44E52", "#8172B2", "#CCB974"]

plt.figure(figsize=(9, 9))
bars = plt.bar(categories, values, color=colors, edgecolor="black")

# Add labels and title
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.title("Fancy Bar Chart with Five Categories", fontsize=14, fontweight="bold")

# Add value labels above bars
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2, 
        yval + 1, 
        f"{yval}", 
        ha="center", 
        va="bottom", 
        fontsize=10, 
        fontweight="bold"
    )

# Show horizontal grid for readability
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
categories = ["Category A", "Category B", "Category C", "Category D", "Category E"]
values = [23, 45, 12, 36, 29]

# Colors for the pie chart
colors = ["#4C72B0", "#55A868", "#C44E52", "#8172B2", "#CCB974"]

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    values, 
    labels=categories, 
    autopct="%1.1f%%",   # show percentages
    startangle=140,      # rotate start
    colors=colors, 
    wedgeprops={"edgecolor": "black"}  # clean edges
)

# Add title
plt.title("Pie Chart with Five Categories", fontsize=14, fontweight="bold")

plt.show()
```
